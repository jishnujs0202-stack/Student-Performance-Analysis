```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student-mat.csv", sep=";")

print("Dataset Loaded Successfully")
df.head()
```

```python
print("Shape of Dataset:", df.shape)

df.info()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
avg_grade = df["G3"].mean()

print("Average Final Grade (G3):", round(avg_grade,2))
```

```python
above_15 = df[df["G3"] > 15].shape[0]

print("Students Scoring Above 15:", above_15)
```

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation Between Study Time and G3:", round(correlation,3))
```

```python
gender_avg = df.groupby("sex")["G3"].mean()

print(gender_avg)
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10, edgecolor="black")

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")

plt.show()
```

```python
gender_avg.plot(kind="bar")

plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")

plt.show()
```
